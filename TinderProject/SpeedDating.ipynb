```python
iddata['income'] = pd.to_numeric(iddata['income'].str.replace(',', ''), errors='coerce')

def income_range(income):
    if pd.isna(income):
        return 'NaN'
    elif income < 10000:
        return '0-10k'
    elif income < 20000:
        return '10k-20k'
    elif income < 30000:
        return '20k-30k'
    elif income < 40000:
        return '30k-40k'
    elif income < 50000:
        return '40k-50k'
    elif income < 60000:
        return '50k-60k'
    elif income < 70000:
        return '60k-70k'
    elif income < 80000:
        return '70k-80k'
    elif income < 90000:
        return '80k-90k'
    elif income < 100000:
        return '90k-100k'
    else:
        return '100k+'

iddata['inc_range'] = iddata['income'].apply(income_range)

income_order = ['NaN', '0-10k', '10k-20k', '20k-30k', '30k-40k', '40k-50k', 
                '50k-60k', '60k-70k', '70k-80k', '80k-90k', '90k-100k', '100k+']


fig_income_range = px.histogram(iddata, x='inc_range', color='gender', barmode='group', 
                                color_discrete_map={0: 'blue', 1: 'red'},
                                labels={'gender': 'Gender', 'inc_range': 'Income Range'},
                                category_orders={'inc_range': income_order, 'gender': [0, 1]},
                                title="Histogram of Income Range grouped by Gender")


fig_income_range.for_each_trace(lambda t: t.update(name=gender_labels[int(t.name)]))

fig_income_range.update_layout(xaxis_title='Income Range', yaxis_title='Count',
                               legend_title_text='Gender')

fig_income_range.show()
```

### Satisfaction regarding test conditions

```python
value_labels = {
    1: 'Too few',
    2: 'Too many',
    3: 'Just right'
}


iddata['length_label'] = iddata['length'].map(value_labels)
iddata['numdat_2_label'] = iddata['numdat_2'].map(value_labels)


plot_data = pd.melt(iddata, id_vars=['gender'], 
                    value_vars=['length_label', 'numdat_2_label'],
                    var_name='Attribute', value_name='Preference')


plot_data['Attribute'] = plot_data['Attribute'].replace({
    'length_label': 'Length',
    'numdat_2_label': 'Number of Dates'
})


fig = px.histogram(plot_data, x='Attribute', color='Preference', pattern_shape="gender", barmode='group', 
                   pattern_shape_sequence=["", "/"], title="Preferences by Gender and Attribute",
                   labels={'Preference': 'Preference Level', 'Attribute': 'Attribute'},
                   color_discrete_map={'Too few': 'blue', 'Too many': 'red', 'Just right': 'green'})


fig.update_layout(xaxis_title='Attribute', yaxis_title='Count', legend_title_text='Preference')
fig.show()
```

#### Using an int_score which the harmonic mean of shar1_1 between participants, we look at how interst match (int_corr) impacted the matchs

```python
partner_data = data.rename(columns={'iid': 'pid', 'pid': 'iid', 'shar1_1': 'partner_shar1_1'})

merged_data = data.merge(partner_data[['iid', 'pid', 'partner_shar1_1']], on=['iid', 'pid'])

merged_data['int_score'] = 2 * merged_data['shar1_1'] * merged_data['partner_shar1_1'] / (merged_data['shar1_1'] + merged_data['partner_shar1_1'])

match_data = merged_data[merged_data['match'] == 1]
no_match_data = merged_data[merged_data['match'] == 0]


match_avg_corr = match_data.groupby('int_score')['int_corr'].mean().reset_index(name='avg_int_corr')
match_avg_corr['Match'] = 'Yes'  

no_match_avg_corr = no_match_data.groupby('int_score')['int_corr'].mean().reset_index(name='avg_int_corr')
no_match_avg_corr['Match'] = 'No' 


plot_data = pd.concat([match_avg_corr, no_match_avg_corr])


fig = px.scatter(plot_data, x='int_score', y='avg_int_corr', color='Match',
              title="Evolution of Average Interest Correlation with Int Score",
              labels={'int_score': 'Int Score (Sum of shar1_1 values)', 'avg_int_corr': 'Average Interest Correlation', 'Match': 'Match'})


fig.update_layout(xaxis_title='Int Score', yaxis_title='Average Interest Correlation')
fig.show()
```

```python
attributes_o = ['attr_o', 'sinc_o', 'fun_o', 'amb_o', 'shar_o']
attribute_means = iddata.groupby(['matchs', 'gender'])[attributes_o].mean().reset_index()
gender_labels = {0: 'Male', 1: 'Female'}
attribute_means['Gender'] = attribute_means['gender'].map(gender_labels)
attribute_means_melted = attribute_means.melt(id_vars=['matchs', 'Gender'], 
                                              value_vars=attributes_o,
                                              var_name='Attribute', 
                                              value_name='Average Value')


coef_data = []


for attribute in attributes_o:
    for gender in gender_labels.values():
        subset = attribute_means_melted[(attribute_means_melted['Attribute'] == attribute) & 
                                        (attribute_means_melted['Gender'] == gender)]
        
        X = subset[['matchs']].values
        y = subset['Average Value'].values
        
        
        model = LinearRegression()
        model.fit(X, y)
        
        
        coef = model.coef_[0]
        intercept = model.intercept_
        
        
        coef_data.append({'Attribute': attribute, 'Gender': gender, 'Coefficient': coef, 'Intercept': intercept})
        
        
        x_range = np.linspace(subset['matchs'].min(), subset['matchs'].max(), 100)
        y_range = model.predict(x_range.reshape(-1, 1))
        
        
        fig.add_trace(go.Scatter(x=x_range, y=y_range, mode='lines', 
                                 name=f"{attribute} - {gender} Regression", 
                                 line=dict(dash='dash'),
                                 legendgroup=f"{attribute} - {gender}"))


coef_df = pd.DataFrame(coef_data)



fig = px.line(attribute_means_melted, x='matchs', y='Average Value', color='Gender', 
              facet_col='Attribute', facet_col_wrap=5,
              title="Average Attribute Value by Match Count and Gender",
              labels={'matchs': 'Match Count', 'Average Value': 'Average Attribute Value'})

fig.update_layout(
    xaxis_title='Match Count', 
    yaxis_title='Average Attribute Value', 
    width=1400, 
    height=400  
)

fig.show()
```
